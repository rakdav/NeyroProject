```python
from keras import datasets,Sequential,layers,utils,preprocessing,models
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
from keras.src.saving import load_model
```

```python
classes =['футболка','брюки','свитер','платье','пальто','туфли',
'рубашка','кроссовки','сумка','ботинки']
```

```python
model=models.load_model('fashion_mnist_dense.h5')
model.summary()
```

```python
img_path='koza_back.jpg'
Image(img_path,width=150,height=150)
```

```python
img=preprocessing.image.load_img(img_path,target_size=(28,28),color_mode='grayscale')
x=preprocessing.image.img_to_array(img)
x=x.reshape((1,784))
x/=255
```

```python
prediction=model.predict(x)
prediction
```

```python
prediction=np.argmax(prediction)
print("Номер класса",prediction)
print("Название класса",classes[prediction])
```
